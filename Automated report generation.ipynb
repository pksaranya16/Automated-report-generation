```python
# Automated Report Generation using Pandas and FPDF

import pandas as pd
from fpdf import FPDF

# Read data from CSV file
df = pd.read_csv("E:/data.csv")

# Analyze data
total_records = len(df)
average_score = df["Score"].mean()
highest_score = df["Score"].max()
lowest_score = df["Score"].min()

# Create PDF report
pdf = FPDF()
pdf.add_page()

# Title
pdf.set_font("Arial", "B", 16)
pdf.cell(200, 10, "AUTOMATED REPORT GENERATION", ln=True, align="C")

pdf.ln(10)

# Summary Section
pdf.set_font("Arial", "B", 12)
pdf.cell(200, 10, "Summary Statistics", ln=True)

pdf.set_font("Arial", "", 12)
pdf.cell(200, 10, f"Total Records: {total_records}", ln=True)
pdf.cell(200, 10, f"Average Score: {average_score:.2f}", ln=True)
pdf.cell(200, 10, f"Highest Score: {highest_score}", ln=True)
pdf.cell(200, 10, f"Lowest Score: {lowest_score}", ln=True)

pdf.ln(10)

# Table Header
pdf.set_font("Arial", "B", 12)
pdf.cell(90, 10, "Name", border=1)
pdf.cell(50, 10, "Score", border=1)
pdf.ln()

# Table Data
pdf.set_font("Arial", "", 12)
for index, row in df.iterrows():
    pdf.cell(90, 10, str(row["Name"]), border=1)
    pdf.cell(50, 10, str(row["Score"]), border=1)
    pdf.ln()

# Save PDF
pdf.output("report.pdf")

print("PDF report generated successfully!")
print("File saved as: report.pdf")


```
